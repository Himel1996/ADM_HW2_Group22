```python
import pandas as pd
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
json_file_path = '/content/drive/My Drive/HW2/lighter_books.json'
```

```python
# Chunk size
chunk_size = 200000
# Initialize an empty list to store chunks
books = pd.DataFrame()
# Loop through each chunk and append it to the result dataframe
for chunk in pd.read_json(json_file_path, lines=True, chunksize=chunk_size):
    # Select only the required columns from the chunk
    selected_chunk = chunk[["num_pages","original_publication_date","title"]]
    #selected_chunk = chunk[["original_publication_date"]]
    # Append the selected chunk to the list
    books = pd.concat([books, selected_chunk], ignore_index=True)


```

```python
books_copy = books
```

```python
books_copy.dropna(subset=['original_publication_date', 'num_pages', 'title'], inplace=True)
```

```python
# Check if "original_publication_date" is in the correct date format
books_copy['original_publication_date'] = pd.to_datetime(books['original_publication_date'], errors='coerce')

# Drop rows with incorrect dates
books_copy = books_copy[books_copy['original_publication_date'].notna()]
```

```python
# Filter out rows with "original_publication_date" greater than the present date
from datetime import datetime
present_date = datetime.now()
books_copy = books_copy[books_copy['original_publication_date'] <= present_date]
books_copy["original_publication_date"].max()
```

```python
# Filter out rows with incorrect values in "num_pages" (e.g., non-numeric values)
books_copy['num_pages'] = pd.to_numeric(books_copy['num_pages'], errors='coerce')
books_copy = books_copy[books_copy['num_pages'].notna()]
books_copy = books_copy[books_copy['num_pages'] != 0]
books_copy["num_pages"].min()
```

```python
books_copy = books_copy[books_copy['num_pages'] >= 50]
```

```python
books_copy['title'] = books_copy['title'].str.strip()
```

```python
df=books_copy
```

```python
def books_info_for_year(year):
    #year = pd.to_datetime(year, format='%Y')
    # Filter the DataFrame to include only books published in the given year
    books_in_year = df[df['original_publication_date'].dt.year == year]

    # Calculate the number of books published in the given year
    Num_Books = len(books_in_year)

    # Calculate the total number of pages written in the given year
    Total_Pages = books_in_year['num_pages'].sum()

    # Convert the publication date to month and count the most prolific month
    books_in_year['publication_month'] = books_in_year['original_publication_date'].dt.month
    #Most_Prolific_Month = books_in_year['publication_month'].value_counts().idxmax()

    # Find the longest book written in the given year
    #longest_book = books_in_year[books_in_year['num_pages'] == books_in_year['num_pages'].max()]['title'].values[0]

    Most_Prolific_Month = None
    longest_book = None

    if not books_in_year.empty:
        # Extract the month from the publication date and find the most prolific month
        books_in_year['month'] = books_in_year['original_publication_date'].dt.month
        Most_Prolific_Month = books_in_year['month'].value_counts().idxmax()

        # Find the title of the longest book written that year
        longest_book = books_in_year[books_in_year['num_pages'] == books_in_year['num_pages'].max()]['title'].values[0]



    return Num_Books, Total_Pages, Most_Prolific_Month, longest_book
```

```python
year = 2003 # Replace with the desired year
result = books_info_for_year(year)
print(result)
```

```python
years = df['original_publication_date'].dt.year.unique()
yearly_info_df = pd.DataFrame(columns=['Year', 'Num_Books', 'Total_Pages', 'Most_Prolific_Month', 'Longest_Book'])
# Loop through the years and calculate the information
for year in years:
  Num_Books, Total_Pages, Most_Prolific_Month, Longest_Book = books_info_for_year(year)
  yearly_info_df = yearly_info_df.append({'Year': year, 'Num_Books': Num_Books, 'Total_Pages': Total_Pages, 'Most_Prolific_Month': Most_Prolific_Month, 'Longest_Book': Longest_Book}, ignore_index=True)
```
